### Diabetes prediction using patient details

```python
# import nessecary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import itertools

import warnings
warnings.filterwarnings('ignore')
```

#### Load DataSet

```python
DATASETPATH = "https://raw.githubusercontent.com/MainakRepositor/Diabetes-Prediction-System/master/diabetes.csv"
```

```python
# Read the dataset
data = pd.read_csv(DATASETPATH)

print("DataSet Loaded Sucessfully....")
```

```python
# Access the dataset
data.head(5)
```

```python
# Dataset dimensions
data.shape # 768 => rows && 9 => cols
```

```python
# Basic meta data about the data
data.info()
```

```python
# Statistical Summary
data.describe().T
```

```python
# Check for missing values
try:
    NullValues = data.isnull().sum()
    print("Null Values Total Count")
    print(NullValues)
except:
    print("Mistake in the Dataset")
```

#### Data Visualizations

```python
sns.countplot(x = 'Outcome', data = data) 
```

```python
# Histogram
cols = data.columns[:8]
length = len(cols)
plt.subplots(figsize = (20, 15))

for i, j in itertools.zip_longest(cols, range(length)):
    plt.subplot(round((length/2)), 3, j+1)
    data[i].hist(bins = 10)
    plt.title(i)
plt.show()
```

```python
# PairPlot
sns.pairplot(data = data, hue = 'Outcome')
plt.show()
```

```python
# HeatMap
sns.heatmap(data = data.corr(), annot=True)
plt.show()
```
